```python
import tensorflow as tf
import tensorflowjs as tfjs

# Load the Keras h5 model
h5_model_path = '/content/drive/MyDrive/캡스톤/품종 딥러닝/dog_model.h5'
keras_model = tf.keras.models.load_model(h5_model_path)

# Convert the Keras model to TensorFlow.js Layers model
tfjs_model_dir = '/content/drive/MyDrive/캡스톤/품종 딥러닝/tfjsmodel'
tfjs.converters.save_keras_model(keras_model, tfjs_model_dir)

print('Model converted successfully!')
```
